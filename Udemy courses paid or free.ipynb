### Data Modeling

#### For Data Modeling we have to split our data into training and test set. We can use this splitted dataset into various models and check the scores of each model. Then, We can select the model with the highest score.



```python
from sklearn.model_selection import train_test_split 
from sklearn import metrics
from sklearn.metrics import classification_report, confusion_matrix
```

```python
from sklearn.model_selection import train_test_split
model=udemy
```

```python
id = model[model['num_lectures'] == 0].index.values[0]
id
```

```python
model.drop(id,axis=0,inplace=True)
```

```python
model.describe().transpose()
```

```python
model.drop(['course_title','url','time'],axis=1,inplace=True)
```

```python
model['is_paid'] = model['is_paid'].apply(lambda x: 1 if x is True else 0 )
```

```python
def sub(val): 
  
    if val == 'Business Finance': 
        return 0
    elif val == 'Graphic Design': 
        return 1
    elif val == 'Musical Instruments': 
        return 2
    else:
        return 3
    
model['subject'] = model['subject'].apply(sub)
```

```python
def year(val): 
  
    if val == 2017: 
        return 6
    elif val == 2016: 
        return 5
    elif val == 2015: 
        return 4
    elif val == 2014: 
        return 3
    elif val == 2013: 
        return 2
    elif val == 2012: 
        return 1
    else:
        return 0
    
model['year'] = model['year'].apply(year)
```

```python
def lev(val): 
  
    if val == 'All Levels': 
        return 0
    elif val == 'Intermediate Level': 
        return 1
    elif val == 'Beginner Level': 
        return 2
    else:
        return 3
    
model['level'] = model['level'].apply(lev)
```

```python
X = model.drop(['course_id','is_paid'],axis=1)
X.head()
```

```python
y = model['is_paid']
y.head()
```

```python
X.shape,y.shape
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=10)
```

```python
X_train.shape,X_test.shape,y_train.shape,y_test.shape
```

### Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
lrmodel = LogisticRegression()
lrmodel.fit(X_train,y_train)
```

```python
lpredictions = lrmodel.predict(X_test)
```

```python
#Accuracy Matrix
print("Confusion Matrix")
print(classification_report(y_test,lpredictions))
print(confusion_matrix(y_test,lpredictions))
```

```python
lscore = lrmodel.score(X_test, y_test)
ltscore = lrmodel.score(X_train, y_train)

print(lscore)
print(ltscore)
```

### K-Nearest Neighbour

```python
from sklearn.neighbors import KNeighborsClassifier
kmodel = KNeighborsClassifier(n_neighbors=3)
kmodel.fit(X_train,y_train)
```

```python
kpredictions = kmodel.predict(X_test)
```

```python
print("Confusion Matrix")
print(confusion_matrix(y_test,kpredictions))

kscore = round((kmodel.score(X_test, y_test)*100),2)
print ("\nModel Score:",kscore,"%")
```
